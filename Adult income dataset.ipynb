# univariate Analysis

Univariate analysis is a statistical method used to analyze and describe the characteristics of a single variable. In other words, it focuses on examining the distribution, central tendency, and dispersion of a single variable without considering its relationship with other variables.

The key objectives of univariate analysis include    :- **

Descriptive Statis** tics: Univariate analysis helps in summarizing the main features of a dataset by providing measures such as mean, median, mode, range, variance, and standard deviat
- **n.

Visuali**zation: Graphical techniques such as histograms, box plots, and pie charts are commonly used in univariate analysis to visualize the distribution and patterns within a single vari- **ble.

Detection of **Outliers: Univariate analysis can identify outliers or extreme values within a dataset that may skew the results or indicate errors in data coll- **ction.

Understanding Data Di**stribution: It helps in understanding the shape and characteristics of the distribution of a variable, such as whether it follows a normal distribution, skewed distribution, or has multip- **e peaks.

Identification of Pattern**s and Trends: Univariate analysis can reveal patterns and trends over time or across different categories within a singl    e   variable.

While univariate analysis provides valuable insights into individual variables, it is often complemented by multivariate analysis, which explores the relationships between multiple variables simultaneously. Multivariate analysis helps in understanding the complex interactions and dependencies among variables, offering a more comprehensive understanding of the data.

# Bivariate Analysis

Bivariate analysis is a statistical method used to explore the relationship between two different variables. Unlike univariate analysis, which focuses on analyzing a single variable at a time, bivariate analysis examines how two variables are related to each other.

The main objectives of bivariate analysis include:

- **Understanding Relationships**: Bivariate analysis helps in understanding the relationship, association, or correlation between two variables. It allows researchers to determine whether changes in one variable are associated with changes in the other variable.
  
- **Exploring Dependencies**: Bivariate analysis helps in exploring dependencies between two variables. It identifies whether changes in one variable are dependent on the values of the other variable.

- **Identifying Patterns and Trends**: Bivariate analysis can reveal patterns and trends that emerge when two variables are considered together. It helps in understanding how the values of one variable vary with the values of another variable.

- **Predictive Modeling**: In predictive modeling and machine learning, bivariate analysis can be used to identify important predictor variables that are strongly associated with the target variable.

- **Visualization**: Graphical techniques such as scatter plots, line plots, and heatmaps are commonly used in bivariate analysis to visualize the relationship between two variables.

- **Hypothesis Testing**: Bivariate analysis can be used to test hypotheses about the relationship between two variables. Statistical tests such as correlation analysis, t-tests, chi-square tests, and ANOVA are often used to assess the significance of the relationship.

*Bivariate analysis forms the foundation for more advanced multivariate analysis techniques, which involve the simultaneous analysis of three or more variables. By understanding the relationship between two variables, researchers can gain insights into complex systems and phenomena, leading to better decision-making and problem-solving in various fields such as economics, social sciences, healthcare, and engineering.*

#### replace salary values['<=50k','>50k'] with 0 and 1

```python
alldata.columns
```

```python
alldata['income'].unique()
```

```python
alldata['income'].value_counts()
```

```python
sns.countplot(x='income', data=alldata)
```

```python
def income_data(inc):
    if inc=='<=50K':
        return 0
    else:
        return 1
```

```python
alldata['encoded_salary']=alldata['income'].apply(income_data)
alldata.head(10)
```

```python
# another method
alldata.replace(to_replace=['<=50K', '>50K'], value=[0,1], inplace=True)
alldata.tail(10)
```
