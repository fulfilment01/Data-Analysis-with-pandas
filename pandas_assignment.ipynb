## Assignment 4

* Using the datasets above: List_of_local_governments_in_Nigeria.csv and Nigeria_povertyrate_per_local_gov.csv
    1. Merge the dataset together to form a single table.
    2. Remove any redundant columns.
    3. Calculate how population affects poverty rate of each local government.
    4. Arrange each local government accordiing to the order of their poverty rate.
    5. Find the percentage of poverty and sort in ascending order.

```python
data1=pd.read_csv("Nigeria data/list_of_local_government_areas_of_nigeria.csv")
data1
```

```python
data2=pd.read_csv("Nigeria data/Nigeria povertyrate per local gov.csv")
data2
```

## removing the useless columns from each dataset

```python
data1.columns
```

```python
data1=data1.drop(columns="Unnamed: 3")
data1
```

```python
data2.columns
```

```python
data2=data2.drop(columns="Unnamed: 10")
data2
```

```python
data1.columns
```

```python
data1[data1['Local Government Area'].isnull()].count()
```

## Merginging the two datas on the column which is common to the two

```python
data = pd.merge(data1, data2, how="outer", left_on="Local Government Area", right_on="NAME_23")
data
```

```python
data.columns
```

```python
data=data.drop(columns="SNo")
data
```

```python
data=data.drop_duplicates()
data
```

```python
data=data.dropna(axis="rows", how="any")
data
```

```python
data.columns
```

```python
data=data.rename(columns={"NAME_0":"COUNTRY", "NAME_12":"STATES", "NAME_23":"LGA", "2014_Popul":"2014_Population", "Poverty_Ra":"Poverty_Rate", "2014_Pover":"2014_Poverty"})
data
```

```python
data=data.drop(columns="STATES")
data
```

### Creating The Poverty_Percentage column

```python
data["Poverty_Percentage"]= data["Poverty_Rate"]*100
data
```

### Rounding up The poverty_percentage column to 2 decimal place

```python
data["Poverty_Percentage"]=data["Poverty_Percentage"].round(2)
data
```

### Aranging the data based on ascending order

```python
poverty_percent = data.copy()
poverty_percent.sort_values(by="Poverty_Percentage", ascending=True, inplace=True)
poverty_percent
```

### Reseting the index

```python
poverty_percent.reset_index(drop=True)
```

```python
poverty_percent.info()
```

```python
# correla= poverty_percent.corr(method='spearman')
correlation = poverty_percent['Poverty_Rate'].corr(poverty_percent['2014_Population'])
correlation
```

    * The corr() method computes the Pearson correlation coefficient by default, which measures the linear relationship between two variables. The correlation coefficient ranges from -1 to 1, where:    
    
1 indicates a perfect positive linear relationship    ,
0 indicates no linear relationship, a    nd
-1 indicates a perfect negative linear relationsh    ip.
Positive correlation means that as one variable increases, the other variable also tends to increase, while negative correlation means that as one variable increases, the other variable tends to decre    a    se.

After obtaining the correlation coefficient, you can interpret the relationship between the Poverty_Rate and 2014_Population variables based on its value.
